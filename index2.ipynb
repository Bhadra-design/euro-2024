```python
def download_match_report(url, proxy):
    try:
        # Create an unverified SSL context
        context = ssl._create_unverified_context()
        proxy_handler = urllib.request.ProxyHandler({
            'http': proxy,
            'https': proxy
        })
        opener = urllib.request.build_opener(proxy_handler, urllib.request.HTTPSHandler(context=context))
        response = opener.open(url)
        html_content = response.read()
        return html_content
    except Exception as e:
        print(f"Proxy {proxy} failed: {e}")
        return None
```

```python
def extract_filename(url):
    parts = url.split('/')[-1].split('-')
    if len(parts) < 3:
        return "match_report.html"  # Fallback filename in case of unexpected URL format
    team1 = parts[0]
    team2 = parts[1]
    date = parts[2]
    competition = "-".join(parts[3:])
    filename = f"{team1}vs{team2}-{date}-{competition}.html"
    return filename
```

```python
def download_all_match_reports(links_file, proxy, folder_name):
    # Create the folder if it doesn't exist
    os.makedirs(folder_name, exist_ok=True)

    with open(links_file, 'r') as file:
        all_match_report_links = file.read().splitlines()

    failed_urls = []

    for link in all_match_report_links:
        print(f"Downloading {link}...")
        content = download_match_report(link, proxy)
        if content:
            filename = extract_filename(link)
            file_path = os.path.join(folder_name, filename)
            # Save the content to an HTML file
            with open(file_path, 'wb') as f:
                f.write(content)
            print(f"Saved to {file_path}")
        else:
            print(f"Failed to download {link}")
            failed_urls.append(link)
        time.sleep(5)
```
